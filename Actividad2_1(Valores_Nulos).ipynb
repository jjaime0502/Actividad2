```python
data=pd.read_csv('Ventas_totales.csv')
data
```

```python
#Conocemos la cantidad de valores nulos que hay por columna
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Conocemos la cantidad total de valores nulos en la base de datos
valores_nulos=data.isnull().sum().sum()
valores_nulos
```

```python
data2=data.copy()
data2.info()
```

##Sustitución de valores nulos

```python
#Sustituir valores nulos con promedio o media
data2["salon_ventas"]=data2["salon_ventas"].fillna(round(data["salon_ventas"].mean(),1))
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos con mediana
data2["tarjetas_credito"]=data2["tarjetas_credito"].fillna(round(data["tarjetas_credito"].median(),1))
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por un valor numérico en concreto
data2["tarjetas_debito"]=data["tarjetas_debito"].fillna(1000)
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por un valor numérico en concreto
data2["otros_medios"]=data["otros_medios"].fillna(26600)
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante "forward all"("ffill")
#Filtro por columna
data2["subtotal_ventas_alimentos_bebidas"]=data2["subtotal_ventas_alimentos_bebidas"].fillna(method="ffill")
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos con mediana
data2["bebidas"]=data2["bebidas"].fillna(round(data["bebidas"].median(),1))
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos con promedio o media
data2["almacen"]=data2["almacen"].fillna(round(data["almacen"].mean(),1))
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por valores no nulos hacia atrás "backward fill"("ffill")
data2["panaderia"]=data2["panaderia"].fillna(method="bfill")
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos con mediana
data2["lacteos"]=data2["lacteos"].fillna(round(data["lacteos"].median(),2))
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante "forward all"("ffill")
#Filtro por columna
data2["carnes"]=data2["carnes"].fillna(method="ffill")
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Segundo método de sustitución de valores nulos
#Sustituir valores nulos con mediana
data2["verduleria_fruteria"]=data["verduleria_fruteria"].fillna(round(data["verduleria_fruteria"].median(),1))
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante "forward all"("ffill")
#Filtro por columna
data2["alimentos_preparados_rotiseria"]=data2["alimentos_preparados_rotiseria"].fillna(method="ffill")
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por un valor numérico en concreto
data2["indumentaria_calzado_textiles_hogar"]=data["indumentaria_calzado_textiles_hogar"].fillna(56660)
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos con promedio o media
data2["electronicos_articulos_hogar"]=data2["electronicos_articulos_hogar"].fillna(round(data["electronicos_articulos_hogar"].mean(),1))
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos con promedio o media
data2["otros"]=data2["otros"].fillna(round(data["otros"].median(),1))
```
